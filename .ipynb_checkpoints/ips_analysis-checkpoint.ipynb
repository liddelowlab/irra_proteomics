```python
%%R

# load required packages
library(DESeq2)
library(sva) 
library(tidyverse)
library(RColorBrewer)
library(ComplexHeatmap) 
library(DEGreport) 
library(tximport) 
library(ggplot2)
library(ggrepel)
library(biomaRt)
library(stringr)
library(readxl)
library(ggpubr) 
library(viridis)
library(rtracklayer)
library(limma)
library(forcats)
```

```python
%%R

# read in RNA-seq sample metadata file (RNA-seq data from Hasel et al 2021, PMID: PMID: 34413515, GEO accession GSE165069)
samples <- read.table("input_data/ips_tici_astros/seq_stats.csv", header = TRUE, sep = ",")
```

```python
%%R

samples
```

```python
%%R
# create and name salmon transcript counts file paths (file tree structure is "relative_file_path/input_data/[Sample]/quant.sf", with subfolders for each samples)
files <- file.path("input_data/ips_tici_astros/salmon/output", samples$Sample.ID, "quant.sf")
names(files) <- samples$Run
```

```python
%%R

files
```

```python
%%R
# read in Ensembl rat genome GTF file (obtained from https://ftp.ebi.ac.uk/pub/databases/gencode/Gencode_human/release_44/gencode.v44.annotation.gtf.gz)
human.gtf.df <- as.data.frame(import('input_data/ips_tici_astros/gencode.v44.annotation.gtf.gz'))

# create transcript-to-gene data frame
tx2gene <- human.gtf.df[,c("transcript_id", "gene_name")]
```

```python
%%R

head(tx2gene)
```

```python
%%R
# drop rows of transcript-to-gene data frame which contain an NA value for either column
tx2gene <- tx2gene %>% drop_na(gene_name) %>% drop_na(transcript_id)
```

```python
%%R

head(tx2gene, n = 25)
```

```python
%%R

tx2gene$transcript_id = gsub(tx2gene$transcript_id, pattern="\\.[[:digit:]]+$", replacement="")
```

```python
%%R

head(tx2gene, n = 25)
```

```python
%%R

# import salmon counts files
txi <- tximport(files, type="salmon", txIn = TRUE, txOut = FALSE, tx2gene=tx2gene, importer=read_tsv, ignoreTxVersion=TRUE)
```

```python
%%R

samples$condition <- gsub(samples$Sample.ID, pattern="-[[:digit:]]+$", replacement="")
samples$condition <- substring(samples$condition, 7)
```

```python
%%R

samples$batch <- gsub(samples$Sample.ID, pattern="-[[:digit:]]+$", replacement="")
samples$batch <- substring(samples$batch, 1, 6)
```

```python
%%R

samples
```

```python
%%R
# create DESeq dataset object
ddsTxi <- DESeqDataSetFromTximport(txi,
                                   colData = samples,
                                   design = ~condition + batch)
```

```python
%%R

dds <- ddsTxi
```

```python
%%R

# estimate size factors, retrieve normalized counts, and filter out genes for which normalized counts are not greater than or equal to 5 in at least two samples
dds <- estimateSizeFactors(dds)
nc <- counts(dds, normalized=TRUE)
filter <- rowSums(nc >= 5) >= 2
ddssva <- dds[filter,]
```

```python
%%R

# create model matrices for sva
dat  <- counts(ddssva, normalized = TRUE)
idx  <- rowMeans(dat) > 1
dat  <- dat[idx, ]
mod  <- model.matrix(~condition + batch, colData(ddssva))
mod0 <- model.matrix(~batch, colData(ddssva))
```

```python
%%R
# run sva
set.seed(111)
svseq <- svaseq(dat, mod, mod0)
```

```python
%%R
# add surrogate variables to DESeq2 design
ddssva$SV1 <- svseq$sv[,1]
ddssva$SV2 <- svseq$sv[,2]
design(ddssva) <- ~ SV1 + SV2 + batch + condition
```

```python
%%R
# estimate size factors, dispersions, and run negative bionmial Wald Test
ddssva <- estimateSizeFactors(ddssva)
ddssva <- estimateDispersions(ddssva)
ddssva <- nbinomWaldTest(ddssva, maxit = 1000)
```

```python
%%R -w 8 -h 6 --units in -r 300
# plot dispersion estimates
plotDispEsts(ddssva)
```

```python
%%R
# obtain rlog data
rld <- rlog(ddssva, blind=FALSE)
rlog.exp <- assay(rld)
```

```python
%%R
# run limma removeBatchEffect accounting for our 'batch' variable
mm <- model.matrix(~ SV1 + SV2 + condition, colData(rld))

batch.corrected.rlog.exp <- limma::removeBatchEffect(rlog.exp, batch=rld$batch, design=mm)
assay(rld) <- batch.corrected.rlog.exp
```

```python
%%R -w 4 -h 4 --units in -r 300
# extract PCA results from plotPCA deseq2 function 
pcaplot <- plotPCA(rld, intgroup=c("condition", "batch"))  
pca.data <- plotPCA(rld, intgroup=c("condition", "batch"), returnData = TRUE)  

# create PCA plot
ggplot(pca.data,aes(x=PC1,y=PC2,color=condition)) + 
  geom_point(aes(shape = batch), size = 3) + 
  scale_color_manual(values = rev(c("#B4464B", "#4682B4"))) +
  theme_pubr() + 
  theme(aspect.ratio = 1, legend.position = "right") + 
  labs(color = "Condition", x = pcaplot$labels$x, y = pcaplot$labels$y) 
```

```python
%%R

# Basic function to convert mouse to human gene names
convertMouseGeneList <- function(x){
require("biomaRt")
#mart <- useEnsembl("ensembl","mmusculus_gene_ensembl", mirror = "useast")
human = useMart("ensembl", dataset = "hsapiens_gene_ensembl", host = "https://dec2021.archive.ensembl.org/")
mouse = useMart("ensembl", dataset = "mmusculus_gene_ensembl", host = "https://dec2021.archive.ensembl.org/")
genesV2 = getLDS(attributes = c("mgi_symbol"), filters = "mgi_symbol", values = x , mart = mouse, martL = human, attributesL = c("hgnc_symbol"), uniqueRows=T)
return(genesV2)
}
```

```python
%%R

ortho.df <- convertMouseGeneList(musGenes)
```

```python
%%R

trimmed.ortho.merge.df <- ortho.merge.df[, c("HGNC.symbol", "Significance", "log2FC")] 
```

```python
%%R

merge.results.df <- merge(results.df, trimmed.ortho.merge.df, by.x = "gene", by.y = "HGNC.symbol")
```

```python
%%R

head(merge.results.df)
```

```python
%%R

merge.results.df <- merge.results.df %>% mutate(select_label = ifelse(Significance == "Both (concordant)(n = 67)",
                                                                      ifelse(log2FC > 0, "Upregulated IRRA marker", "Downregulated IRRA marker"), "Not IRRA marker"))
merge.results.df <- merge.results.df %>% mutate(p = -log10(padj)) %>% mutate(p = ifelse(padj == 0 , -log10(.Machine$double.xmin), -log10(padj)))
```

```python
%%R

merge.results.df <- merge.results.df %>% mutate(delabel = ifelse(Significance == "Both (concordant)(n = 67)", gene, NA))
```

```python
%%R

merge.results.df$select_label <- factor(merge.results.df$select_label, levels = c("Upregulated IRRA marker", "Downregulated IRRA marker", "Not IRRA marker"))
```

```python
%%R

merge.results.df <- merge.results.df %>% mutate(order = ifelse(select_label == "Not IRRA marker",
                                                               1, ifelse(select_label == "Upregulated IRRA marker", 2,
                                                                         ifelse(select_label == "Downregulated IRRA marker", 3, 0))))
```

```python
%%R -w 8 -h 4 --units in -r 300

plot <- ggplot(data = merge.results.df %>% arrange(order), aes(x = log2FoldChange, y = p, col = select_label, label = delabel, order = order)) +
  geom_vline(xintercept = c(-0.2, 0.2), col = "gray", linetype = 'dashed') +
  geom_hline(yintercept = -log10(0.05), col = "gray", linetype = 'dashed') +
  geom_point(size = 1) +
  scale_color_manual(values = rev(c("lightgray", "darkred", rev(RColorBrewer::brewer.pal(n=3, name = "Dark2"))[3])))  + # to set the labels in case we want to overwrite the categories from the dataframe (UP, DOWN, NO)
    ggpubr::theme_pubr() + theme(legend.position = "right") + 
    scale_y_continuous(limits = c(0, 330), expand = c(0, 0)) + 
    labs(x = "log2-Fold Change", y = "-log10(adjusted p-value)", color = NULL) + 
    geom_text_repel(max.overlaps = Inf, size = 2, show.legend = FALSE) # To show all labels 
    
plot
```
